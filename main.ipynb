```

import face_recognition

%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib.patches as Rectangle
import matplotlib.patches as Circle
import numpy as np
import cv2

##image database
image1 = cv2.imread('cr1.jpeg')
cr = cv2.cvtColor(image1, cv2.COLOR_BGR2RGB)
cr_encoding = face_recognition.face_encodings(cr)[0]

image2 = cv2.imread('m1.jpeg')
m = cv2.cvtColor(image2, cv2.COLOR_BGR2RGB)
m_encoding = face_recognition.face_encodings(m)[0]

image3 = cv2.imread('nr1.jpeg')
nr = cv2.cvtColor(image3, cv2.COLOR_BGR2RGB)
nr_encoding = face_recognition.face_encodings(nr)[0]

image4 = cv2.imread('zi1.jpeg')
zi = cv2.cvtColor(image4, cv2.COLOR_BGR2RGB)
zi_encoding = face_recognition.face_encodings(zi)[0]

image4 = cv2.imread('r1.jpeg')
r = cv2.cvtColor(image4, cv2.COLOR_BGR2RGB)
r_encoding = face_recognition.face_encodings(r)[0]

known_face_encodings = [cr_encoding, m_encoding, nr_encoding, zi_encoding, r_encoding]

image = cv2.imread('mix6.jpeg')             #here you need to specify the name of your .jpeg file to recognise players
unknown_image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.axis('off')
plt.imshow(unknown_image)

unknown_face_encondings = face_recognition.face_encodings(unknown_image)

from scipy.spatial import distance

for unknown_face_encoding in unknown_face_encondings:
  results = []

  for known_face_encoding in known_face_encodings:
    d = distance.euclidean(unknown_face_encoding, known_face_encoding)
    results.append(d)

  threshold = 0.6
  results = np.array(results) <= threshold

  name = 'Unknown'

  if results[0]:
    name = 'Ronaldo'

  if results[1]:
    name = 'Messi'

  if results[2]:
    name = 'Neymar'

  if results[3]:
    name = 'Ibrahimović'

  if results[4]:
    name = 'Ronaldinho'

  print('Person is {}'.format(name))  # At present it  can indentify Messi, Neymar, Ronaldo, Ronaldinho and Ibrahimović
```
